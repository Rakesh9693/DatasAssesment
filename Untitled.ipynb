# Missing Data
#Missing Data - Initial Intuition.
#Here, we don't have any missing data.
#Therefore we can use the dataset without working on missing data.


```python
#Create a copy of base data for manupulation & processing
#Total Charges should be numeric amount. Let's convert it to numerical data type
telco_data = telco_base_data.copy()
telco_data.TotalCharges = pd.to_numeric(telco_data.TotalCharges, errors='coerce')
telco_data.isnull().sum()
```

```python
#since there are 11 missing values in TotalCharges column.we can check them by
telco_data.loc[telco_data ['TotalCharges'].isnull() == True]
```

```python
#Missing Value Treatement

#Since the % of these records compared to total dataset is very low ie 0.15%, it is safe to ignore them from further processing.
telco_data.dropna(how = 'any', inplace = True)
```

```python
#Remove columns not required for processing
telco_data.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
# checking the coloumn again whether it is updated or not
telco_data.head()
```

# Data Exploration
**1. ** Plot distibution of individual predictors by churn


```python
#Univariate Analysis
#Ploting grapg camparing all the required coloumns in datasets
for i, predictor in enumerate(telco_data.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges']).columns):
    plt.figure(i)
    sns.countplot(data=telco_data, x=predictor, hue='Churn', palette='inferno')
    plt.title(f'Count plot of {predictor} by Churn')
    plt.xlabel(predictor)
    plt.ylabel('Count')
    #plt.legend(title='Churn')
    plt.xticks(rotation=45)
    plt.tight_layout()  
    plt.show()
```


# Insight we got From This Univariate Analysis
High Churn Rate in Month-to-Month Contracts as compared from 1yrs and 2 yrs.
In Paymaent method coustomer are churning more in Electronic check method as compared to others etc


# Convert the target variable 
'Churn' in a binary numeric variable i.e. Yes=1 ; No = 0
so that we can use them as numeric value


```python
telco_data['Churn'] = np.where(telco_data.Churn == 'Yes',1,0)

telco_data.head()
```

# Convert all the categorical variables into dummy variables


```python
telco_data_dummies = pd.get_dummies(telco_data)
telco_data_dummies.head()
```

```python
 #Churn by Monthly Charges and Total Charges
```

```python
Mth = sns.kdeplot(telco_data_dummies.MonthlyCharges[(telco_data_dummies["Churn"] == 0) ],
                color="green", shade = True)
Mth = sns.kdeplot(telco_data_dummies.MonthlyCharges[(telco_data_dummies["Churn"] == 1) ],
                ax =Mth, color="Blue", shade= True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly charges by churn')
```

```python
Tot = sns.kdeplot(telco_data_dummies.TotalCharges[(telco_data_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Tot = sns.kdeplot(telco_data_dummies.TotalCharges[(telco_data_dummies["Churn"] == 1) ],
                ax =Tot, color="Blue", shade= True)
Tot.legend(["No Churn","Churn"],loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total charges by churn')
```

# INSIGHT
Higher Monthly Charge at lower tenure results into lower Total Charge. Hence, all these 3 factors viz Higher Monthly Charge, Lower tenure and Lower Total Charge are linkd to High Churn.

# Build corelation of all predictors withChurn

```python
plt.figure(figsize=(20, 10))

# Calculate the correlation with the 'Churn' column and sort the values
corr_with_churn = telco_data_dummies.corr()['Churn'].sort_values(ascending=False)

# Plot the correlations as a bar plot
ax = corr_with_churn.plot(kind='bar', color="pink")

# Add title and labels with appropriate font sizes
plt.title('Correlation of Features with Churn', fontsize=20)
plt.xlabel('Features', fontsize=20)
plt.ylabel('Correlation Coefficient', fontsize=20)

# Rotate x-axis labels for better readability and adjust font size
plt.xticks(rotation=90, fontsize=11)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=1)
plt.axhline(0, color='black', linewidth=0.2)
plt.tight_layout()
plt.show()
```

# Derived Insight:
HIGH Churn seen in case of Month to month contracts, No online security, No Tech support, First year of subscription and Fibre Optics Internet

LOW Churn is seens in case of Long term contracts, Subscriptions without internet service and The customers engaged for 5+ years

Factors like Gender, Availability of PhoneService and # of multiple lines have alomost NO impact on Churn

##This is also evident from the Heatmap below

```python
plt.figure(figsize=(25, 20))
corr_matrix = telco_data_dummies.corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, cmap="Paired", annot=True, fmt=".2f", linewidths=.5, cbar_kws={"shrink": .5})
plt.title('Correlation Matrix Heatmap', fontsize=25)
plt.xlabel('Features', fontsize=20)
plt.ylabel('Features', fontsize=20)
plt.xticks(fontsize=15, rotation=90)
plt.yticks(fontsize=15)
plt.show()
```

These are some of the insights from this dataset:-
